```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv(r"C:\Users\halwa\OneDrive\Desktop\Algo\data. csv")
```

```python
data
```

```python
data = data.set_index(pd.DatetimeIndex(data['Date'].values))
```

```python
plt.figure(figsize=(16,8))
plt.title('Close Price')
plt.plot(data['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.show()
```

```python
period1 = 20
period2 = 66

# SMA

data['SMA1'] = data['Close'].rolling(window = period1).mean()
data['SMA2'] = data['Close'].rolling(window = period2).mean()
```
